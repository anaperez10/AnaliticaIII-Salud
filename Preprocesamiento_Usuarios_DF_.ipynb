```python
#Convertir columnas en minúsculas
users.columns= users.columns.str.lower()
users.head(5)
```

```python
#Renombrar columnas 
users.rename(columns = {'year':'año'},inplace=True)
users.head(3)
```

```python
#Tipo de variables
users.dtypes
```

```python
#Imprimir categorias de las variables
lista = ['mes', 'sexo', 'fecha nacimiento',
       'departamento', 'municipio', 'barrio', 'fecha inicio al pgp',
       'fecha primera clase funcional', 'primera clase funcional',
       'fecha última clase funcional', 'última clase funcional', 'ciclo_vital',
       'quinquenio']
for i in lista:
    print(users[i].value_counts())
    print('------------------------------------')
```

```python
#Aplicar funciones para homologar a nivel general.
lista = ['mes', 'sexo', 'fecha nacimiento',
       'departamento', 'municipio', 'barrio', 'fecha inicio al pgp',
       'fecha primera clase funcional', 'primera clase funcional',
       'fecha última clase funcional', 'última clase funcional', 'ciclo_vital',
       'quinquenio']
for i in lista:
    users[i] = users[i].str.lower() #minuscula
    users[i] = users[i].str.strip() #espacios
    users[i] = users[i].apply(lambda x: normalize("NFKD", x).encode("ascii", "ignore").decode("ascii") if pd.notnull(x) else x)
```

```python
#Seleccionar solo los datos de femenino y masculino
users1 = users[(users['sexo']=='f') | (users['sexo']=='m')]
users1['sexo'].value_counts()
```

```python
#Modificar formato fecha
#'fecha última clase funcional'
users1['fecha última clase funcional'] = users1['fecha última clase funcional'].apply(lambda x: str(x)[0:10]) 
users1['fecha última clase funcional'] = pd.to_datetime(users1['fecha última clase funcional'])
```

```python
#Modificar formato fecha
#'fecha primera clase funcional'
users1['fecha primera clase funcional'] = users1['fecha primera clase funcional'].apply(lambda x: str(x)[0:10]) 
users1['fecha primera clase funcional'] = pd.to_datetime(users1['fecha primera clase funcional'])
```

```python
#Modificar formato fecha
#'fecha inicio al pgp'
users1['fecha inicio al pgp'] = users1['fecha inicio al pgp'].apply(lambda x: str(x)[0:10]) 
users1['fecha inicio al pgp'] = pd.to_datetime(users1['fecha inicio al pgp'])
```

```python
#Modificar formato fecha
#'fecha nacimiento'
users1['fecha nacimiento'] = users1['fecha nacimiento'].apply(lambda x: str(x)[0:10]) 
users1['fecha nacimiento'] = pd.to_datetime(users1['fecha nacimiento'])
```

```python
#Observar el tipo de los datos despues de pasar todas las fechas a formato fecha
users1.dtypes
```

```python
users2 = users1[['mes', 'año', 'nrodoc', 'sexo', 'edad',
                 'municipio', 'fecha inicio al pgp','fecha primera clase funcional',
                 'primera clase funcional','fecha última clase funcional', 
                 'última clase funcional','ciclo_vital','quinquenio']]
users2.head(5)
```

```python
#Pasar los nulos de usuarios a sin información
users2['municipio'] = users2['municipio'].replace(np.nan,'sin información')
```

```python
#Eliminar datos que no incluye la edad 
users2 = users2.dropna(subset=['edad'])
```

```python
#Convertir la edad en entero
users2['edad'] = users2['edad'] = users2['edad'].astype(int)
```
