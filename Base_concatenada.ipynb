```python
cronicos = cronicos.drop(['Unnamed: 0','Tipo','Clasificación IMC','Calificación (Índice de Fragilidad)','Calificación (Apoyo Monopodal)','Calificación Velocidad'], axis=1)
cronicos
```

```python
cronicos.columns= cronicos.columns.str.lower()
cx.columns= cx.columns.str.lower()
```

```python
usuarios = usuarios.drop(['Unnamed: 0','ciclo_vital','quinquenio'],axis=1)
```

```python
#Base de datos con la informacion de todos los empleados
df = pd.merge(usuarios,cx, on = 'nrodoc',how = 'inner')
df1 = pd.merge(df,cronicos, on = 'nrodoc',how = 'inner')
df1
```

```python
df1=df1.drop_duplicates()
```

```python
df1['codigo servicio habilitacion']=df1['codigo servicio habilitacion'].fillna(217.0)
```

```python
df1['tipo de anestesia']=df1['tipo de anestesia'].fillna('sin informe quirurgico')
```

```python
df2=df1[['nrodoc','tipo procedimiento','fecha cirugia']]
df2
```

```python
df2.dtypes
```

```python
df2['fecha cirugia']=pd.to_datetime(df2['fecha cirugia'])
```

```python
df2['Year'] = df2['fecha cirugia'].dt.year
df2['Mes'] = df2['fecha cirugia'].dt.month
```

```python
df2
```

```python
df2=df2.astype({'Mes': object})
```

```python
import plotly.express as px
```

```python
# crear dataset
base = df2.groupby(['Mes', 'Year'])[['nrodoc']].count().reset_index()
# MSSubClass: Tipo de edificio
# YrSold: Año en que se vende
# SalePrice: Precio de venta

# crear gráfica
fig = px.bar(base, x = 'Year', y='nrodoc', color = 'Mes', barmode = 'group', title= '<b>Precio de venta según el tipo de deficio por año<b>')

# agregar detalles a la gráfica
fig.update_layout(
    xaxis_title = 'Año',
    yaxis_title = 'Precio de venta',
    template = 'simple_white',
    title_x = 0.5)

fig.show()
```
